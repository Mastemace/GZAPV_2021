```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,20)
```

```python
img = cv.imread('img.jpg')
```

```python
kernel = np.ones((5,5))
kernel /= np.sum(kernel)

out_avg   = cv.filter2D(img, -1, kernel)
out_gauss = cv.GaussianBlur(img, (5,5), 1)
out_med   = cv.medianBlur(img, 5)
out_bil   = cv.bilateralFilter(img, 5, 150, 150)
out_nl    = cv.fastNlMeansDenoisingColored(img, 5)
```

```python
num_imgs = 6

plt.subplot(1,num_imgs,1)
plt.imshow(img)
plt.subplot(1,num_imgs,2)
plt.imshow(out_avg)
plt.subplot(1,num_imgs,3)
plt.imshow(out_avg)
plt.subplot(1,num_imgs,4)
plt.imshow(out_med)
plt.subplot(1,num_imgs,5)
plt.imshow(out_bil)
plt.subplot(1,num_imgs,6)
plt.imshow(out_nl)
plt.show()
cv.imwrite("avg.jpg",out_avg)
cv.imwrite("gauss.jpg",out_gauss)
cv.imwrite("med.jpg",out_med)
cv.imwrite("bil.jpg",out_bil)
cv.imwrite("nl.jpg",out_nl)
```
