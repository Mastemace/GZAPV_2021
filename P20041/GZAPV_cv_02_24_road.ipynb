```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv.imread('road.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
canny = cv.Canny(gray,255,255)
lines = cv.HoughLinesP(canny, 1, np.pi/180, 310, minLineLength=100, maxLineGap=150)

for line in lines:
  x1, y1, x2, y2 = line[0]
  cv.line(img, (x1, y1), (x2, y2), (255,0,0), 5)

cv.imwrite('road_lines.jpg', img)
```

```
plt.figure(figsize=(20,20))
plt.imshow(img)
```
