```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv.imread('sample/image5.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray, 300, 400)

#plt.figure(figsize=(20,20))
#plt.imshow(edges)
lines = cv.HoughLinesP(edges, 1, np.pi/180, 250, minLineLength=250, maxLineGap=300)

for line in lines:
  x1, y1, x2, y2 = line[0]
  cv.line(img, (x1, y1), (x2, y2), (255, 0, 0), 3)

plt.figure(figsize=(20,20))
plt.imshow(img)
#cv.imwrite('sample2/out_image_result.jpg', img)
```
