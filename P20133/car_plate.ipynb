```
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
#  Это хорошо работает на русских машинах 

img = cv2.imread('supra3.jpg')

rohlik = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_russian_plate_number.xml')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#plt.imshow(img)

plates = rohlik.detectMultiScale(img, scaleFactor=1.25, minNeighbors=3)

for (x, y, w, h) in plates:
  img = cv2.rectangle(img, (x,y), (x+w, y+h), (0, 0, 255), 3)
cv2.imwrite('res.png', img)
res = cv2.imread('res.png')

plt.figure(figsize=(10,10))
plt.imshow(res)
```
